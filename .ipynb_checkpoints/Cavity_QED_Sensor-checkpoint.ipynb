## Cavity-coupled quantum electrodynamic (QED) sensor
 Hanfeng Wang, Matthew E. Trusheim, and Dirk Englund - Dec. 2022

Resonators for electromagnetic fields in the range of 1-100 GHz are essential in a range of quantum technologies, including superconducting quantum computers, microwave-optical quantum information interfaces, as well as sensors. Here, we consider microwave cavities coupled to ensembles of electron and nuclear spins associated with nitrogen vacancy (NV) centers in diamond. Following the discovery of coherent quantum dynamics of NV centers in the late 1990s, NV centers have emerged as leading quantum sensors for quantities ranging from electromagnetic fields to temperature, strain, rotation (gyroscopes), and have even been proposed for exotic schemes like dark matter detection. The key advantages of NV centers in sensing applications, over other electron spin systems, are their long coherence times and avaible optical polarization at room temperature.

Upon optical excitation, nonradiative decay through a spin-state-dependent intersystem crossing produces spin-state-dependent fluorescence contrast and polarize NV spin to m_s = 0 ground state. A common way for NV spin readout relies on detecting this spin-state-dependent fluorescence. However, due to the low photon detection efficiency (C~0.015) for NV ensemble, the NV magnetic sensitivity is limited by the photon-shot-noise-limit. 

One way to solve this problem is to use NV-cavity coupling, where we couple NV centers to another system (cavity in this case) and read the NV resonance frequency by via the cavity mode.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import HBox, VBox
from IPython.display import display
%matplotlib inline

@widgets.interact_manual(
    g0_in=(0.01, 1.),kc0_in=(1., 1000.),kc1_in=(1., 1000.),ks_in=(1., 1000.))
def runFunc(g0_in=0.01, kc0_in = 70, kc1_in = 111, ks_in = 308):
    kc0 = 2*np.pi*kc0_in; #kHz  Cavity linewidth
    kc1 = 2*np.pi*kc1_in; #kHz   Cavity-loop coupling (output coupling)

    ks = 2*np.pi*ks_in; #kHz  Spin coherence time


    density_NVs = 5; #PPM Density of NV centers in the diamond, given in ppm
    diamondDensity = 1.76e29; # atoms/m^3 Atomic density of diamond
    density_NVs_SI = density_NVs*1e-6*diamondDensity; # NV/m^3

    diamond_sample_dim_x=3e-3; # m Diamond spatial dimensions, assuming cuboid
    diamond_sample_dim_y=3e-3; # m
    diamond_sample_dim_z=0.5e-3; # m
    diamond_sample_volume = diamond_sample_dim_x*diamond_sample_dim_y*diamond_sample_dim_z; 

    N_NVs= density_NVs_SI*diamond_sample_volume ; # the number of NVs in the diamond

    g0 = g0_in*1e-3; #kHz   Coupling strength for single spin, see referenced papers  

    geff = g0*np.sqrt(N_NVs); # Total coupling scales as sqrt(N)


    kc = kc0 + kc1; # Total cavity loss

    xnum = 101; # number of points to plot in x
    ynum = 121;

    # initialize sweep parameters 
    wdc = np.linspace(-0.8,0.8,num=xnum)*1e3*2*np.pi;  #kHz  Detuning of the probe frequency and cavity frequency
    wsc = np.linspace(-7.4,7.4,num=ynum)*1e3*2*np.pi;  #kHz  Detuning of the spin frequency and cavity frequency

    # initialize output ararys
    inphaseOut = np.zeros((xnum,ynum));
    quadratureOut = np.zeros((xnum,ynum));

    for ii in range(xnum):

      for jj in range(ynum):

        wds = wdc[ii]-wsc[jj]; #spin-cavity detuning

        # Cavity reflection signal from references
        refOutput = -1 + kc1/(kc/2+1j*wdc[ii]+pow(geff,2)/(ks/2+1j*wds)); 
        inphaseOut[ii][jj] = np.real(refOutput); # In-phase part of the cavity signal 
        quadratureOut[ii][jj] = np.imag(refOutput);  # Quadrature part of the cavity signal 

    y, x = np.meshgrid(np.linspace(-7.4,7.4,num=ynum), np.linspace(-0.8,0.8,num=xnum),)
    fig, figAxes = plt.subplots(2,2,sharex='all', sharey='all')
    ax1 = figAxes[0,0];
    ax2 = figAxes[0,1];
    ax3 = figAxes[1,0];
    ax4 = figAxes[1,1];


    c = ax1.pcolormesh(y, x, quadratureOut)
    ax1.set_title('Q')
    #ax1.set_xlabel('fsc (MHz)')
    ax1.set_ylabel('fdc (MHz)')
    fig.colorbar(c, ax=ax1)

    c = ax2.pcolormesh(y, x, inphaseOut)
    ax2.set_title('I')
    #ax2.set_xlabel('fsc (MHz)')
    #ax2.set_ylabel('fdc (MHz)')
    fig.colorbar(c, ax=ax2)

    c = ax3.pcolormesh(y, x, np.sqrt(pow(quadratureOut,2)+pow(inphaseOut,2)))
    ax3.set_title('P')
    ax3.set_xlabel('fsc (MHz)')
    ax3.set_ylabel('fdc (MHz)')
    fig.colorbar(c, ax=ax3)

    c = ax4.pcolormesh(y, x, np.arctan( np.divide(quadratureOut,inphaseOut)))
    ax4.set_title('Phi')
    ax4.set_xlabel('fsc (MHz)')
    #ax4.set_ylabel('fdc (MHz)')
    fig.colorbar(c, ax=ax4)

    plt.show()      
```

Here is the sensitivity analysis for cavity-coupled sensor. In the ideal case, the spin-projection-limit sensitivity is propotional to the inverse of the square root of the number of spins (N). However, the NV decoherence time and dephasing time will decrease during the increase of the NV number. Therefore, the sensivitity will decrease sharply in the large-N regime. The script below calculated the spin-projection-limit sensitivity and photon-shot-noise-limit with different number of spins.

The script below also include the cavity-coupled spin sensor. The scale of the sensitivity in the low-N regime is linear with the 1/N rather than 1/sqrt(N), which offers the possibility for this sensor to go below the spin-projection-limit sensitivity.

```python
import numpy as np
import matplotlib.pyplot as plt

Delta = 2e6;
f = 0.1;

Sphi = 1e-9; # phase noise

Q = 3e4;  # Quality factor

hbar = 1.05e-34;  

gs = 2; 

muB = 9.27e-24; 

rho = np.logspace(8,20,40)*1e6;

Tc = 1/3.3e-19/(rho*(0.95/0.05));

T2c = 300e-6;

T2 = 1/(1/T2c+1/Tc);

mu0 = 4*np.pi*1e-7;

C = 0.015;

etaB = 8 * pow((hbar/gs/muB),3) * Delta/np.pi/Q/T2 * Sphi * 1/mu0/hbar/rho *1/f * np.exp(pow((T2/T2c),3))* np.exp(T2/Tc)* np.sqrt(1+1e-6/T2);

eta = hbar/muB/gs * np.pi * np.exp(pow((T2/T2c),3))/C/np.sqrt(rho*T2) * np.exp(T2/Tc) * np.sqrt(1+1e-6/T2);

eta_proj = hbar/muB/gs * np.pi * np.exp(pow((T2/T2c),3))/np.sqrt(rho*T2) * np.exp(T2/Tc) * np.sqrt(1+1e-6/T2);

fig,ax = plt.subplots()

ax.loglog(rho/1e6,eta*1/np.sqrt(3e-3*3e-3*0.5e-3),label='Photon shot noise limit')

ax.loglog(rho/1e6,eta_proj*1/np.sqrt(3e-3*3e-3*0.5e-3),label='Spin projection limit')

ax.loglog(rho/1e6,etaB,label='Cavity readout sensitivity')

ax.legend()

ax.set_xlabel('NV Density (cm^{-3})')
ax.set_ylabel('Sensitivity (T/\sqrt(Hz))')


```

```python
def compute_sens(rho, gammao, V, Q, gamma, PowerList, vot, Vd, energy, gs, kc, kc1, Gamma=None):
    if Gamma==None:
        Gamma = 330e3 / 4 * rho
    alpha2 = 10**(PowerList / 10) * 1e-3 / (energy * kc1)
    beta2 = alpha2 * kc1
    N = (rho * 3.51 / 12.01 * 6.022e23) * Vd * min(Vd,V)/Vd
    eta = 1*(1+0.7)*(1/6)/(1*0.7)
    I_1 = 3*(1+0.7)*(2+1)*200/0.7
    Polarization = (1-eta)/(1 + 2*eta + I_1/gammao)
    geff = gs * np.sqrt(N * Polarization)   # correct
#     geff = 0.19e6*np.sqrt(rho) * np.sqrt(min(V,Vd)/Vd) * np.sqrt(Vd/(3e-3*3e-3*0.9e-3))
    gamma1 = gamma + gammao
    xi1 = np.sqrt(4 * 2 * gs**2 * alpha2 / gammao * gamma1 + gamma1**2)
    alpha2_v2 = kc1 * beta2 / (kc / 2 + geff**2 * 2 * gamma1 / xi1 / (Gamma + xi1))**2
    xi3 = np.sqrt(4 * 2 * gs**2 * alpha2_v2 / gammao * gamma1 + gamma1**2)
    dVdB = geff**2 * 4 / (Gamma + xi3)**2 * kc1 / (kc / 2 + geff**2 * 2 * gamma1 / xi3 / (Gamma + xi3))**2 * vot * 28e9 / np.sqrt(3) * 1e-4 * 2                
    xi = xi1
    xi2 = gamma1 / np.sqrt(xi**2 + gamma1**2)
    geff2 = geff * np.sqrt(xi2)
    Gamma1 = Gamma + gamma1 / xi2
    CC = kc1 * Gamma1 * geff2**2 / (Gamma1**2 / 4 * kc**2 / 4 + geff2**2 * Gamma1 * kc / 2 + geff2**4)
    noise = 0.45e-9 - 0.1345e-9 * CC
    return min(noise / dVdB / 1e4)
```

```python
from scipy.integrate import quad

def compute_sens2(rho, gammao, V, Q, gamma, PowerList, vot, w, L, energy, gs, kc, kc1, alpha, Gamma=None):
    if Gamma==None:
        Gamma = 330e3 / 4 * rho
    alpha2 = 10**(PowerList / 10) * 1e-3 / (energy * kc1)
    beta2 = alpha2 * kc1
#     geff = 0.19e6 * gs / 0.019 * np.sqrt(V / Vd) * np.sqrt(rho / 4)
    eta = 1*(1+0.7)*(1/6)/(1*0.7)
    I_1 = 3*(1+0.7)*(2+1)*200/0.7
    def spin_pol_density(l):
        gammao_l = gammao*np.exp(-alpha*l)
        P = (1-eta)/(1 + 2*eta + I_1/gammao_l)
        return (rho * 3.51 / 12.01 * 6.022e23) * w**2 * P
    N, error = quad(spin_pol_density,0,L)
    Vd = w*w*L
    geff = gs * np.sqrt(N * min(Vd,V)/Vd) # correct
    gamma1 = gamma + gammao
    xi1 = np.sqrt(4 * 2 * gs**2 * alpha2 / gammao * gamma1 + gamma1**2)
    alpha2_v2 = kc1 * beta2 / (kc / 2 + geff**2 * 2 * gamma1 / xi1 / (Gamma + xi1))**2
    xi3 = np.sqrt(4 * 2 * gs**2 * alpha2_v2 / gammao * gamma1 + gamma1**2)
    dVdB = geff**2 * 4 / (Gamma + xi3)**2 * kc1 / (kc / 2 + geff**2 * 2 * gamma1 / xi3 / (Gamma + xi3))**2 * vot *28e9 / np.sqrt(3) * 1e-4 * 2                
    xi = xi1
    xi2 = gamma1 / np.sqrt(xi**2 + gamma1**2)
    geff2 = geff * np.sqrt(xi2)
    Gamma1 = Gamma + gamma1 / xi2
    CC = kc1 * Gamma1 * geff2**2 / (Gamma1**2 / 4 * kc**2 / 4 + geff2**2 * Gamma1 * kc / 2 + geff2**4)
    noise = 0.45e-9 - 0.1345e-9 * CC
    return min(noise / dVdB / 1e4)
```

### Sensitivity Optimization Over Diamond Volume and NV Density

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

def power_linear(rho, V, Vd, P_per_NV):
    N = (rho * 3.51 / 12.01 * 6.022e23) * V0 * min(V0,V)/V0
    return N*P_per_NV

P_per_NV = 5/(3e-3*3e-3*0.9e-3*(4 * 3.51 / 12.01 * 6.022e23))
# P_per_NV = 5/(1e14)

rho_list = np.logspace(-3, 5, 501)
GammaList = 330e3 * rho_list  # inhomogeneous linewidth (Hz)
gammaoList = np.linspace(10, 100, 21) * 1e3  # optical polarization rate
gamma = 4e3   # homogeneoud linewidth

PowerList = np.linspace(-55, 100, 51)
vot = np.sqrt(10**(PowerList / 10) * 1e-3 * 50)

r1 = np.logspace(-3.5, 3, 501)
Vd = 3e-3 * 3e-3 * 0.9e-3   # diamond volume
V0list = Vd * r1   
mu0 = 4 * np.pi * 1e-7
energy = 6.626e-34 * 2.87e9   # h * omega
kc0 = 130e3   # cavity linewidth
kc1 = 130e3   # cavity loop coupling
Q = 2e4  # Quality factor
kc = kc0 + kc1  # cavity loss
gs = 0.019   # single coupling strength  

V = 1 * energy / 3 / gs**2 * (28e9)**2 * mu0  # 2/3 due to the geometry factor

sens_Vd_rho = np.zeros((len(V0list), len(rho_list)))
powers_lin = np.zeros((len(V0list), len(rho_list)))

for ii in range(len(r1)):
    V0 = V0list[ii]   # diamond volume
    for jj in range(len(rho_list)):
        rho = rho_list[jj]
        sens0 = []
        
        for kk in range(len(gammaoList)):
            gammao = gammaoList[kk]     # optical polarization rate
            Gamma = GammaList[jj]   # inhomogeneous linewidth
            power = power_linear(rho, V, Vd, P_per_NV)
            sens1 = compute_sens(rho, gammao, V, Q, gamma, PowerList, vot, V0, energy, gs, kc, kc1, Gamma=Gamma)
            sens0.append((sens1, power))
        min_sens = min(sens0, key=lambda t: t[0])
        sens_Vd_rho[ii, jj] = min_sens[0]
        powers_lin[ii,jj] = min_sens[1]
```

### Sensitivity Optimization over Cavity Mode Volume and Quality Factor

```python
from scipy.optimize import shgo
import numpy as np
import pandas as pd

def sens_wrapper(params, V, Q, gamma, PowerList, vot, w, L, energy, gs, kc, kc1):
    rho, gammao = params    # optimize over NV density and optical polarization rate
    alpha = 0.45 * 100 * rho/0.1
    return compute_sens2(rho, gammao, V, Q, gamma, PowerList, vot, w, L, energy, gs, kc, kc1, alpha)
    
Qlist = np.logspace(2, 6, 101)   #quality factor
gamma = 4e3   #homogeneous linewidth

PowerList = np.linspace(-70, 20, 51)
vot = np.sqrt(10**(PowerList / 10) * 1e-3 * 50)
Vd = 3e-3 * 3e-3 * 0.9e-3   #diamond volume
Vlist = Vd * np.logspace(-6, 4, 101)   #cavity mode volume
mu0 = 4 * np.pi * 1e-7   
energy = 6.626e-34 * 2.87e9   #h*omega
rhomax = 0.1231 * 4 * 1e-6 / Vd

sens_V_Q = np.zeros((len(Vlist), len(Qlist)))
opt_params_V_Q = np.zeros((len(Vlist), len(Qlist)), dtype=object)

for ii in range(len(Vlist)):
    V = Vlist[ii]
    gs = np.sqrt(energy * (28e9)**2 * mu0 / 3 / V)   #single coupling strength

    for jj in range(len(Qlist)):
        Q = Qlist[jj]   # Quality factor
        kc0 = 2.87e9 / Q   # cavity linewidth
        kc1 = kc0    # cavity loop coupling
        kc = kc0 + kc1   # cavity loss
        
        initial_guess = (1,50e3)
        bounds = ((1e-2,rhomax),(10e3,100e3))
        solution = shgo(sens_wrapper, bounds, args=(V, Q, gamma, PowerList, vot, 3e-3, 0.9e-3, energy, gs, kc, kc1))
        sens_V_Q[ii, jj] = solution.fun
        opt_params_V_Q[ii, jj] = solution.x
```

```python
def min_power_linear(max_sens, V, Q, gammao, gamma, PowerList, vot, energy, gs, kc, kc1, Gamma=Gamma):
    # Locate the minimum power that can achieve some sensitivity value max_sense or better based on linear scaling
    # of optical power with number of NVs
    
    P_per_NV = 5/(3e-3*3e-3*0.9e-3*(4 * 3.51 / 12.01 * 6.022e23))
    # P_per_NV = 5/(1e14)
    rholist = np.logspace(-2,2,50)
    Vdlist = np.logspace(-3,3,50) * 3e-3*3e-3*0.9e-3
    powers = []
    Vds = []
    rhos = []
    for i in range(len(Vdlist)):
        Vd=Vdlist[i]
        for j in range(len(rholist)):
            rho=rholist[j]
            rhomax = 0.1231 * 4 * 1e-6 / Vd
            if rho>rhomax:
                continue
            sensi = compute_sens(rho, gammao, V, Q, gamma, PowerList, vot, Vd, energy, gs, kc, kc1,  Gamma=Gamma)
            if sensi < max_sens:
                power = P_per_NV * Vd * (rho * 3.51 / 12.01 * 6.022e23) 
                powers.append(power)
                Vds.append(Vd)
                rhos.append(rho)
    if not powers:
        return [None, None, None]
    min_index = powers.index(min(powers))
    return [powers[min_index], Vds[min_index], rhos[min_index]]
            
    
```

#### A more rigorous method

```python
from scipy.optimize import minimize, shgo
import numpy as np
import pandas as pd

def ternary_search(max_sens, bounds, gammao, V, Q, gamma, PowerList, vot, L, energy, gs, kc, kc1, alpha, Gamma=None, display=False):
    # Find the minimum diamond width that gives the desired sensitivity and return the power with that value
    minn,maxx=bounds
    rhomax = alpha * 0.1 /100 / 0.45   # max rho allowable by length of diamond (assuming diamond is at its max length)
    # Check if minimum diamond width is sufficient
    rho_minw = min(0.1231 * 4 * 1e-6 / (L * minn**2),100,rhomax)
    if compute_sens2(rho_minw, gammao, V, Q, gamma, PowerList, vot, minn, L, energy, gs, kc, kc1, alpha, Gamma=Gamma) < max_sens:
        power = minn * minn * (6.626e-34*(3e8/532e-9))*gammao/absorp_cs
        return power, [L, minn, rhomin, gammao]
    
    log_min=np.log10(minn)
    log_max=np.log10(maxx)
    sub_max = False
    for i in range(10):
        log_mid1 = log_min + (log_max - log_min) / 3
        log_mid2 = log_max - (log_max - log_min) / 3

        mid1 = 10**log_mid1
        mid2 = 10**log_mid2

        Vd1 = L * mid1**2
        rho1 = min(0.1231 * 4 * 1e-6 / Vd1, 100,rhomax)
        sens1 = compute_sens2(rho1, gammao, V, Q, gamma, PowerList, vot, mid1, L, energy, gs, kc, kc1, alpha, Gamma=Gamma)

        Vd2 = L * mid2**2
        rho2 = min(0.1231 * 4 * 1e-6 / Vd2, 100,rhomax)
        sens2 = compute_sens2(rho2, gammao, V, Q, gamma, PowerList, vot, mid2, L, energy, gs, kc, kc1, alpha, Gamma=Gamma)

        if sens1 > max_sens and sens2 > max_sens:
            # Both midpoints are above the threshold, move the bounds inward
            log_min = log_mid1
            log_max = log_mid2
        elif sens1 <= max_sens and sens2 <= max_sens:
            # Both midpoints are below the threshold, move the right bound inward
            sub_max=True
            log_max = log_mid2
        elif sens1 <= max_sens:
            # Only mid1 is below the threshold
            sub_max=True
            log_max = log_mid2
        else:
            # Only mid2 is below the threshold
            sub_max=True
            log_min = log_mid1
    if not sub_max:
        return np.inf, [0,0,0,0]
    optimal_log_w = (log_min + log_max) / 2
    optimal_w = 10**optimal_log_w
    rho = min(0.1231 * 4 * 1e-6 / (L * optimal_w**2),100,rhomax)
    power = optimal_w * optimal_w * (6.626e-34*(3e8/532e-9))*gammao/absorp_cs
    return power, [L, optimal_w, rho, gammao]
    

def minimize_power(max_sens, gamma, omega, V=None, Q=None, Gamma=None, kc0=None, kc1=None):
    # Find the minimum power to reach some sensitivity value or better
    # Either for one example with a specific mode volume and Q, or for a range of Vs and Qs
    
    Qlist = np.logspace(2, 6, 101)   #quality factor

    PowerList = np.linspace(-70, 20, 51)
    vot = np.sqrt(10**(PowerList / 10) * 1e-3 * 50)
    Vd_ref = 3e-3 * 3e-3 * 0.9e-3   #diamond volume for reference
    Vlist = Vd_ref * np.logspace(-6, 4, 101)   #cavity mode volume
    mu0 = 4 * np.pi * 1e-7   
    energy = 6.626e-34 * omega   #h*omega

    absorp_cs = 2e-17 / (100**2)  # NV absorption cross section 
#     alpha = 0.45 * 100  # absorption constant of diamond at 532nm for a given NV density
    I_1 = 3*(1+0.7)*(2+1)*200/0.7   # critical optical excitation rate for polarization
    I_min = (6.626e-34*(3e8/532e-9))*I_1/absorp_cs
    
#     gammaoList = 2 * np.pi * np.linspace(10, 100, 21) * 1e3  # optical polarization rate
    gammaoList = np.linspace(10, 100, 21) * 1e3
    
    if V!=None and Q!=None:  # if mode volume and quality factor are known and provided
        gs = np.sqrt(energy * (28e9)**2 * mu0 / 3 / V)   #single coupling strength
        kc = kc0 + kc1   # cavity loss
        powers1 = []
        paramss1 = []
        for i in range(len(gammaoList)): 
            gammao = gammaoList[i]
            Llist = np.logspace(-5,-2,21)
            powers2 = []
            paramss2 = []
            for j in range(len(Llist)):
                L = Llist[j]
                I_in = (6.626e-34*(3e8/532e-9))*gammao/absorp_cs
                alpha = -(1/L)*np.log(I_min/I_in)
                bounds = (1e-5,1e-1)
                power, params = ternary_search(max_sens, bounds, gammao, V, Q, gamma, PowerList, vot, L, energy, gs, kc, kc1, alpha, Gamma=Gamma)
                if power!= np.inf:
                    powers2.append(power)
                    paramss2.append(params)
            if powers2:
                min_index = powers2.index(min(powers2))
                powers1.append(powers2[min_index])
                paramss1.append(np.array(paramss2[min_index]))
        
        if powers1:
            min_index = powers1.index(min(powers1))
            min_power = powers1[min_index]
            min_params = paramss1[min_index]
#             print(min_power)
#             print(min_params)
            return min_power, np.concatenate((min_params, np.array([V,Q])))
        else:
            return np.inf, np.array([0,0,0,0,V,Q])

    sens_args = np.zeros((len(Vlist), len(Qlist)), dtype=object)
    opt_power = np.zeros((len(Vlist), len(Qlist)))
    
    for ii in range(len(Vlist)):
        V = Vlist[ii]
        gs = np.sqrt(energy * (28e9)**2 * mu0 / 3 / V)   #single coupling strength
        for jj in range(len(Qlist)):
            Q = Qlist[jj]   # Quality factor
            kc0 = omega / Q   # cavity linewidth
            kc1 = kc0    # cavity loop coupling
            kc = kc0 + kc1   # cavity loss
            powers1 = []
            paramss1 = []
            bounds = (1e-5,1e-1)
            for k in range(len(gammaoList)): # loop through optical polarization rate values
                gammao = gammaoList[k]
                Llist = np.logspace(-5,-2,21)
                powers2 = []
                paramss2 = []
                for l in range(len(Llist)):  # loop through diamond length values
                    L = Llist[l]
                    I_in = (6.626e-34*(3e8/532e-9))*gammao/absorp_cs
                    alpha = -(1/L)*np.log(I_min/I_in)
                    power, params = ternary_search(max_sens, bounds, gammao, V, Q, gamma, PowerList, vot, L, energy, gs, kc, kc1, alpha, Gamma=Gamma)
                    if power!= np.inf:
                        powers2.append(power)
                        paramss2.append(params)
                if powers2:
                    min_index = powers2.index(min(powers2))
                    powers1.append(powers2[min_index])
                    paramss1.append(paramss2[min_index])
            if powers1:
                min_index = powers1.index(min(powers1))
                min_power = powers1[min_index]
                min_params = np.array(paramss1[min_index])
#                 print(min_power)
#                 print(min_params)
                opt_power[ii, jj] = min_power
                sens_args[ii, jj] = np.concatenate((min_params, np.array([V,Q])))
            else:
                opt_power[ii, jj] = np.inf

    return opt_power, sens_args
```

### Permanent Magnet Fluctuations

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

# Estimate rms fluctuations in the magnetic field in a small gap between infinite permeability yokes from a 
# permanent magnet
# Based on theory developed in "Fluctuations of the magnetic field generated by permanent magnets" by Ernst Schloemann, 1984

H_e = 240/10   # effective magnetic field strength inside magnet (T) (given as 240kOe in paper)
V = 1       # permanent magnet volume (cm^3)
muB = 9.274e-24  # Bohr magneton (J/T)
g = 2   # spectroscopic splitting factor
kB = 1.380649e-23  # J/K
T = 300  # room temp 
kBT = kB*T
S = 1/2   # atomic spin
N = 8e22   # number of atoms in magnet for 1cm3 of SmCo5
M0_Ms =  1.09   # saturation magnetization at 0 temp / time average of magnetic moment per unit volume

a = (g*muB*H_e)/(kBT)   # divide H_e by 10 to convert from kOe to Tesla

fluc = np.sqrt(0.67/(S*N) * np.sqrt(1/a) * (M0_Ms)*(M0_Ms-1) )
print(fluc)

def fluctuations(V, H_e):
    a = (g*muB*H_e)/(kBT)   # divide H_e by 10 to convert from kOe to Tesla
    return np.sqrt((1/V) * 0.67/(S*N) * np.sqrt(1/a) * (M0_Ms)*(M0_Ms-1) )

print(fluctuations(1,24))

VList = np.logspace(-2,3,101)
HeList = np.logspace(-3,2,101)

Vgrid,Hgrid = np.meshgrid(VList,HeList)

flucs = fluctuations(Vgrid,Hgrid)

# Plot the heatmap using pcolormesh
plt.figure(figsize=(10, 8))
plt.pcolormesh(Vgrid, Hgrid, flucs, shading='auto', norm=LogNorm(), cmap='hot')

# Add colorbar and label it
cbar = plt.colorbar()
cbar.set_label('Fluctuations')

# Set scales and labels
plt.xscale('log')
plt.yscale('log')
plt.xlabel(f'Magnet Volume (cm$^3$)')
plt.ylabel('Internal Field Strength (T)')

# Show the plot
plt.show()
```

### Stealth Magnet

##### Flux Concentration
Below is an example of a cavity in which the reentrance cones are filled with ferromagnetic material for flux concentration. The idea here would be to achieve some magnification of the external magnetic field in the gap between the reentrance cones where the sensor is placed to raise the minimally detectable magnetic field.

With the normal scale cavity and a field in the x direction, the enhancement factor in the gap between cones is roughly 3. This can likely be increased significnalty with a smaller gap width and more ideal geometry of the flux concentrators, but it's unlikely to reach an enhancement factor that makes the effort worth the extra noise involved.

```python
import numpy as np
import matplotlib.pyplot as plt

enhancement = 3
omega_B = np.logspace(-1,4,100)
mag_noise = 1.4e-12 *(omega_B**(-1/2)) / enhancement
resistivity = 1.7e-8
mu0 = 4 * np.pi * 1e-7
copper_t = 200e-6 + 1e-3
skin_depth = np.sqrt(2*resistivity/(mu0*omega_B*2*np.pi))
B_r = np.exp(-copper_t/skin_depth) * enhancement

fig, ax1 = plt.subplots()
color = 'tab:blue'
ax1.set_xlabel('Field Frequency (Hz)')
ax1.set_ylabel(r'Thermal Magnetization Noise fT/$\sqrt{Hz}$', color=color)
ax1.plot(omega_B, mag_noise, color=color)
ax1.set_yscale('log')

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel(r'$B_{r}/B_{ext}$', color=color)
ax2.plot(omega_B, B_r, color=color)
ax2.set_yscale('log')

fig.tight_layout() 
plt.xscale('log')  
plt.show()
```

High permeability materials used as flux concentrators suffer from thermal matnetization noise that is strongest at low frequencies. Data gathered in Fescenko et al. "Diamond Magnetometer Enhanced by Ferrite Flux Concentrators" gives a crude estimate of this noise, which they discover scales with frequency as $f^{-1/2}$ (blue line). There is also the consideration of how much the field in the sensor location is reduced by the skin effect (red curve). The metallic cavity will prevent high frequency AC fields from reaching the sensor, so as frequency increases, the factor by which the field is reduced increases. The degree of field reduction is dependent on the cavity material, and the thickness of the metallic cavitiy's walls and layer coating the ferrite materials in the reentrance cones. This suggests that a relatively narrow band of frequencies would need to be identified where magnetization noise is not limiting but also where the skin effect doesn't prohibit an enhancement of the field to be sensed in the region of the sensor.

##### Permanent Magnets

We can make use of the skin effect to disguise permanent magnets in the cavity inside the reentrance cones. The magnets will fill the tapered reentrances shown in the cavity below and be coated by a thin layer of metallic material. This allows the permanent magnets to provide a DC field that penetrates through the metal of the cavity to create the bias field for large Zeeman splitting of the NV system at the location of the diamond in the bottom of the gap between reentrances. The metal coating over the magnets ensures the cavity mode is not disrupted by the magnets. This large Zeeman effect creates a high spin transition frequency that necessitates a larger cavity mode frequency, which can be achieved with a scaled down verison of the cavity.

```python
from IPython.display import display, Latex
B_bias = 0.5
omega_s_p = 2.87e9 + 28e9*B_bias/np.sqrt(3)
omega_c_full = 2.12e9
omega_c_scaled = omega_s_p
scale_factor = omega_c_full/omega_c_scaled
V_scaled = 5.717e-8 * (scale_factor**3)
display(Latex(f"With a bias field of 0.5 T, the target frequency is roughly ${omega_c_scaled/1e9:.2f}$ GHz. "
                f"This results in a cavity scaled down by a factor of {scale_factor:.3f} and a mode volume "
                f"of roughly {V_scaled:.3e} $m^{3}$."))
```
